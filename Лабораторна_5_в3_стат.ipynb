```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gamma
n = 9
p = 0.5


gamma_rvs =gamma.rvs(n, loc=0,scale=p, size=1000)
list_=[('x',gamma_rvs), ('exp', np.exp(gamma_rvs+(1/gamma_rvs))), ('log', np.log(1+(1/gamma_rvs))), ('sigma', (1/(1+np.exp(gamma_rvs))*(1-(1/(1+np.exp(gamma_rvs)))))), ('x*sin(x**2)', gamma_rvs*np.sin(gamma_rvs**2)), ('sin(x**2)', np.sin(gamma_rvs**2))]
#print(list_[4])

for i in list_:
    print(i[0])
    M=(1/1000)*np.sum(i[1])
    for j in range(1, 4):
        print(j, 'нецентральний для', i[0], (1/1000)*np.sum(i[1]**j))
        print(j, 'центральний для', i[0], (1/1000)*np.sum((i[1]-M)**j))
i=list_[-1]
M=(1/1000)*np.sum(i[1])
M_03=(1/1000)*np.sum((i[1]-M)**3)
M_04=(1/1000)*np.sum((i[1]-M)**4)
D=((1/1000)*np.sum(i[1]**2))-(M**2)
sigma=np.sqrt(D)
print(M,sigma)
print('коефіцієнт варіації', M/sigma)
print('коефіцієнт асиметрії', M_03/(sigma**3))
print('коефіцієнт ексцесу', (M_04/(sigma**4))-3)
```
